```python
import pixellib
from pixellib.semantic import semantic_segmentation
```

```python
segment_video = semantic_segmentation()
```

```python
segment_video.load_pascalvoc_model("deeplabv3_xception_tf_dim_ordering_tf_kernels.h5")
```

```python
segment_video.process_video_pascalvoc("data/vidSeg/dance.mp4", overlay = True, frames_per_second = 15, output_video_name="dance_seg.mp4")
```

```python
# to capture live video camera 

import cv2
capture = cv2.VideoCapture(0)
segment_video.process_video_pascalvoc(capture, overlay = True, frames_per_second = 15, output_video_name="dance_seg.mp4")
```
